```python
def getNutrientValue(item, nutrient):
  return nutrient_data[nutrient][item]
```

```python
def score(diet):
  # First calculate nutrition %
  nutrients = ["calcium","cholestrol","choline","copper","iron","lipids","magnesium","manganese","panto_acid","phosphorous","potassium","protein","thiamin","vit b12","vit c","vit d"]
  nutrient_req_values = [1300,2300,300,35,150,28,18,420,16,5,1250,4700,2300,900,2,90]
  diet_nut = np.empty(len(diet),len(nutrients))
  diet_nut_percentage = np.empty(len(diet))

#For each item, calculate the percentage of nutrient requirement fulfilled
  for i in range(len(diet_nut)):
    for j in range(0,len(nutrients)):
      diet_nut[item][i] += getNutrientValue(item[0], i) / nutrient_req_values[i]
    diet_nut_percentage[i] = np.mean(diet_nut, axis=1)

  return np.mean(diet_nut_percentage)
```
